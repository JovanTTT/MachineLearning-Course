```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\jovan\\Desktop\\MachineLearning-Course\\MachineLearning-Course\\Data\\StudentsPerformance.csv')
```

```python
df
```

```python
sns.jointplot(data = df, x = 'math score', y = 'reading score', kind = 'hex')
```

```python
sns.jointplot(data = df, x = 'math score', y = 'reading score', kind = 'kde', shade = True)
```

```python
sns.jointplot(data = df, x = 'math score', y = 'reading score', hue = 'gender')
```

```python
sns.pairplot(data = df, hue = 'gender', diag_kind = 'hist', corner = True) #automatically grabbing numerical values from table
```
