```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'level of education', palette = 'Dark2'); 
#colors the plot by hue column
#https://matplotlib.org/tutorials/colors/colormaps.html palette
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200, alpha = 0.2);  #alpha = transparency
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200, style = 'level of education', hue = 'level of education');

plt.savefig('C:\\Users\\jovan\\Desktop\\MachineLearning-Course\\MachineLearning-Course\\Data\\myplot.jpg')
```
