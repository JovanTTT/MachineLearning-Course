```python
a = np.linspace(0, 10, 11)
```

```python
b = a ** 4
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 4), dpi = 200)

axes[(0, 0)].set_xlabel('X axis')
axes[1][1].set_title('1.1 title')
# axes[0].set_ylim(0, 14)
axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

# axes[1].plot(a, b)

fig.suptitle('Figure title', fontsize = 16)

fig.subplots_adjust(wspace = 0.9, hspace = 0.9)

# fig.set_figwidth(10)

fig.savefig('C:\\Users\\jovan\\Desktop\\MachineLearning-Course\\MachineLearning-Course\\Data\\new_subplots.png', bbox_inches = 'tight')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 1)

for ax in axes:
    ax.plot(x, y)

plt.tight_layout()
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3)

for i in range(0, 3):
    for j in range(0, 3):
        axes[i][j].plot(x, y)
        axes[i][j].set_title('Title {}.{}'.format(i, j))
        axes[i][j].set_xlim(0, 6)
        axes[i][j].set_ylim(0, 12)

fig.tight_layout()

fig.savefig('C:\\Users\\jovan\\Desktop\\MachineLearning-Course\\MachineLearning-Course\\Data\\3x3forSubplot.png')
```
