```python
X = pd.get_dummies(df.drop('species', axis = 1), drop_first = True)
```

```python
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
base_preds = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, base_preds))
```

```python
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
```

```python
model.feature_importances_
```

```python
X.columns
```

```python
pd.DataFrame(index = X.columns, data = model.feature_importances_, columns = ['Feature Importance']).sort_values('Feature Importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (12, 8), dpi = 200)
plot_tree(decision_tree = model, feature_names = X.columns, filled = True); 
# X[2] --> 2 represents index of the column, in my case it is 3rd column, flipper, thats why
# flipper has the biggest importance in the dataframe above (feature_names = X.columns fixes this)
# value = [105, 5, 83] --> 105 values of one species, 5 values of another species and 83 from third one
```

```python
len(X_train) # All training samples are entering root node
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test, model_preds))
    print('\n')
    plt.figure(figsize = (12, 8), dpi = 200)
    plot_tree(decision_tree = model, feature_names = X.columns, filled = True); 
```

```python
report_model(model)
```

```python
pruned_tree = DecisionTreeClassifier(max_depth = 3)
```
